# scraping data from wekipedia


```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
wikiurl = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

page = requests.get(wikiurl)

soup = BeautifulSoup(page.text, 'html')
```

```python
largest_companies = soup.find_all('table')[1]
# soup.find('table',class_="wikitable sortable")
largest_private_companies = soup.find_all('table')[2]
companies_by_profit = soup.find_all('table')[3]
```

```python
# find all titles
#largest_companies.find_all('th')
lc_titles = [title.text.strip() for title in largest_companies.find_all('th')]
# print(lc_titles)

#make it the columns

df = pd.DataFrame(columns = lc_titles)
df

```

```python
# find each data
whole_list = largest_companies.find_all('tr')
# check data structure
# largest_companies.find_all('tr')[1]

# for row in table
for row in whole_list[1:]:
    data_list = [data.text.strip() for data in row.find_all('td')]
    df.loc[len(df)]= data_list

df
```

```python
#similar for another 2 table

# soup.find_all('table')[2]
lpc = pd.DataFrame(columns = [lpc_titles.text.strip() for lpc_titles 
                           in largest_private_companies.find_all('th')])
for lpc_row in largest_private_companies.find_all('tr')[1:]:
    lpc_data_list = [lpc_datas.text.strip() for lpc_datas 
                     in lpc_row.find_all('td')]
    lpc.loc[len(lpc)] = lpc_data_list

lpc
```

```python
# soup.find_all('table')[2]
cbp = pd.DataFrame(columns = [cbp_titles.text.strip() for cbp_titles 
                           in companies_by_profit.find_all('th')])
for cbp_row in companies_by_profit.find_all('tr')[1:]:
    cbp_data_list = [cbp_datas.text.strip() for cbp_datas 
                     in cbp_row.find_all('td')]
    cbp.loc[len(cbp)] = cbp_data_list

cbp
```
